```python
def hitung_genap(arr):
    count = 0
    for x in arr:
        if isinstance(x, int):       # memastikan elemen integer
            if x % 2 == 0:           # cek genap
                count += 1
    return count

# Contoh penggunaan:
data = [1, 2, 4, 7, 10]
print(hitung_genap(data))  # Output: 3
```

```python
def nilai_terbesar(arr):
    if not arr:           # cek array kosong
        return None

    maksimum = arr[0]
    for x in arr:
        if x > maksimum:
            maksimum = x
    return maksimum

# Contoh
print(nilai_terbesar([5, 2, 9, 1]))  # Output: 9
```

```python
def balik_string(s):
    return s[::-1]

# Contoh
print(balik_string("halo"))  # Output: "olah"
```

```python
def jumlah_elemen(arr):
    total = 0
    for x in arr:
        total += x
    return total

# Contoh
print(jumlah_elemen([3, 1, 2]))  # Output: 6
```

```python
def cek_palindrome(s):
    return s == s[::-1]

# Contoh
print(cek_palindrome("level"))  # True
print(cek_palindrome("hello"))  # False
```

```python
def hitung_vokal(s):
    vokal = "aiueoAIUEO"
    count = 0
    for ch in s:
        if ch in vokal:
            count += 1
    return count

# Contoh
print(hitung_vokal("informatika"))  # Output: 5
```

```python
def cari_target(arr, target):
    for x in arr:
        if x == target:
            return True
    return False

# Contoh
print(cari_target([4, 7, 1, 9], 7))  # True
```
